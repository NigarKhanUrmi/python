```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)
```

```python
df= pd.read_csv(r'C:\Users\nigar\OneDrive\Desktop\python\Portfolio\movie\movies.csv\movies.csv')
```

```python
df.head()
```

```python
#missing data finding
for col in df.columns:
    pct_missing= np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#fill null with 0
df=df.fillna(0)
```

```python
df.dtypes
```

```python
#change columns type
df['budget']=df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
df['votes']= df['votes'].astype('int64')
df['runtime']= df['runtime'].astype('int64')
```

```python
df.tail()
```

```python
df['release_new']= df['released'].astype('string')
```

```python
df.dtypes
```

```python
new=  df['release_new'].str.split(", ", n = 1, expand = True)
 
# making separate first name column from new data frame
df['new_year']= new[1]
```

```python
year_only= df['new_year'].str.split(' ', n=1, expand=True)
df['year_only']= year_only[0]
```

```python
df= df.sort_values(by=['gross'], inplace= False, ascending=False)
```

```python
companylist= df['company'].unique()
```

```python
companylist[0:2385]
```

```python
#drop_duplicates
df['company'].drop_duplicates().sort_values(ascending= False)
```

```python
plt.scatter(x=df['budget'], y= df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget of Movies in millions')
plt.ylabel('Gross Earning in millions')
plt.show()
```

```python
#plot budget vs gross using seaborn

sns.regplot(x='budget', y= 'gross', data= df, scatter_kws={"color": "red"},line_kws={"color": "blue"})
```

```python
corrmeth= df.corr(method= 'pearson') #pearson, kendall, spearman
```

```python
sns.heatmap(corrmeth, annot= True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
df_numeric= df
for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtype=='object'):
        df_numeric[col_name]=df_numeric[col_name].astype('category')
        df_numeric[col_name]=df_numeric[col_name].cat.codes
  
```

```python
df_numeric.dtypes
```

```python
df_numeric = df_numeric.drop('release_new', axis=1)
```

```python
df_numeric = df.drop('new_year', axis=1)
```

```python
correlation_mat= df_numeric.corr(method='pearson')
sns.heatmap(correlation_mat, annot= True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
#unstacking
corr_mat= df_numeric.corr()
corr_pairs = corr_mat.unstack()
```

```python
sorted_pairs= corr_pairs.sort_values(ascending=False)
sorted_pairs
```
